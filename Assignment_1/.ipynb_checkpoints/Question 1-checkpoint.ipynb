# list of numbers which are multiple of 3 and 7 between 1 to 500

```python
l=[]
counter=0;
for i in range(1,500):
    if (i % 3 == 0 or i % 7 == 0):
        l.insert(counter,i)
        l.sort()
for x in l:print(x)
```

# Mean of the above list

```python
l=[]
counter=0;
for i in range(1,500):
    if (i % 3 == 0 or i % 7 == 0):
        l.insert(counter,i)
        l.sort()
def mean(numbers):
    return float(sum(numbers)) / max(len(numbers), 1)
mean(l)
```

# 25th quartile, 50th quartile and 75th quartile of the above list

```python
import numpy
l=[]
counter=0;
for i in range(1,500):
    if (i % 3 == 0 or i % 7 == 0):
        l.insert(counter,i)
        l.sort()
first=numpy.percentile(l, 25)
second=numpy.percentile(l, 50)
third=numpy.percentile(l, 75)
print("25th Quartile:",first)
print("50th Quartile:",second)
print("75th Quartile:",third)
```
