```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
import glob
import csv

#declare a set to hold all the user ids
user_ids=set()
files=glob.glob('C:/Fall 2016/Data Analysis using Python/Python-Assignments/Midterm/UserData/*.json')   
for file in files:
    with open(file) as f:
        data=f.read()
        user_data=json.loads(data)
        for key in user_data:
            #read all the keys in json and assign it to set 
            user_ids.add(user_data[key]['owner']['user_id'])
            
#convert the set to string
users=str(user_ids)
#make it a string of ids seperated by comma
users = users.replace(", ",";")
users = users.replace("{","")
users = users.replace("}","")
users = users.replace("'","")
results = requests.get(url="https://api.stackexchange.com/2.2/users/"+users+"?site=stackoverflow&key=Qbw)9y9ZFui61wxBzFU7mQ((") 
raw_data=results.json() 

import json
with open('C:/Fall 2016/Data Analysis using Python/Python-Assignments/Midterm/Questions/raw_json.json', 'w') as outfile:
    json.dump(raw_data,outfile)

d={}
for data in raw_data['items']:
    d[data['user_id']]=data

with open('C:/Fall 2016/Data Analysis using Python/Python-Assignments/Midterm/averageBadge.json', 'w') as outfile:
    json.dump(d,outfile)

data_file = open('averageBadge.json','r')
from datetime import datetime,timedelta

def months_between(start,end):
    print(end.year)
    print(start.year)
    print(12 * end.year-12 * start.year)
    diff = (12 * end.year + end.month) - (12 * start.year + start.month)
    return diff

#read the json file
data=json.loads(data_file.read())
badges_count={}
count=0
from datetime import datetime
import time
from time import mktime
for d in data:
    count=data[d]['badge_counts']['gold']+data[d]['badge_counts']['silver']+data[d]['badge_counts']['bronze']
    start=datetime.fromtimestamp(mktime(time.gmtime(data[d]['creation_date'])))
    end = datetime.now()
    months=months_between(start,end)
    averagePerMonth=count/months
    badges_count[d]=averagePerMonth


import texttable as tt
tab = tt.Texttable()
header = ['User ID','Average badges per month']
tab.header(header)
for b in badges_count:
    row = [b,badges_count[b]]
    tab.add_row(row)
s = tab.draw()
print(s)
```
